```python
import pandas as pd
```

```python
stable = [5,5,5,5,5,5,5]

noisy = [1,9,2,8,3,7,2]

transition = [1,1,1,1,8,8,8,8]
```

```python
stable_df = pd.DataFrame({"value": stable})

noisy_df = pd.DataFrame({"value": noisy})

transition_df = pd.DataFrame({"value": transition})
```

```python
stable_df
```

```python
stable_df["rolling_mean"] = stable_df["value"].rolling(window=3).mean()

noisy_df["rolling_mean"] = noisy_df["value"].rolling(window=3).mean()

transition_df["rolling_mean"] = transition_df["value"].rolling(window=3).mean()
```

```python
stable_df
```

```python
noisy_df
```

```python
transition_df
```

```python
plt.figure(figsize=(8,4))

plt.plot(stable_df["value"], label="Original")
plt.plot(stable_df["rolling_mean"], label="Rolling Mean")

plt.legend()
plt.savefig("../images/stable_rollingmean.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,4))

plt.plot(noisy_df["value"], label="Original")
plt.plot(noisy_df["rolling_mean"], label="Rolling Mean")

plt.legend()
plt.savefig("../images/noisy_rollingmean.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,4))

plt.plot(transition_df["value"], label="Original")
plt.plot(transition_df["rolling_mean"], label="Rolling Mean")

plt.legend()
plt.savefig("../images/transition_rollingmean.png", dpi=300)
plt.show()
```

```python
stable_df["rolling_std"] = stable_df["value"].rolling(window=3).std()

noisy_df["rolling_std"] = noisy_df["value"].rolling(window=3).std()

transition_df["rolling_std"] = transition_df["value"].rolling(window=3).std()
```

# Noise and RandomWalk



```python
import random
import matplotlib.pyplot as plt

noise = []

for _ in range(100):
    noise.append(random.choice([-1,1]))

position = 0
walk = [position]

for _ in range(100):
    position += random.choice([-1,1])
    walk.append(position)

plt.figure(figsize=(8,4))

plt.plot(noise,label="Noise")
plt.plot(walk,label="Random Walk")

plt.legend()
plt.savefig("../images/Noise_and_RandomWalk.png", dpi=300)
plt.show()
```
