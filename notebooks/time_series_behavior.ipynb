## Multi-lag autocorrelation (ACF)

Autocorrelation at a single lag only shows immediate memory. Plotting ACF across multiple lags reveals how persistence decays over time, stable and transition systems should show slow decay, noise and random signals should decay to near-zero almost immediately.

```python
from statsmodels.graphics.tsaplots import plot_acf
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(12,8))

plot_acf(stable_df["value"], ax=axes[0,0], title="Stable ACF")
plot_acf(noisy_df["value"], ax=axes[0,1], title="Noisy ACF")
plot_acf(transition_df["value"], ax=axes[1,0], title="Transition ACF")
plot_acf(random_df["value"], ax=axes[1,1], title="Random ACF")

plt.tight_layout()
plt.savefig("../images/acf_comparison.png", dpi=300)
plt.show()
```

# Noise and RandomWalk

Random walks accumulate random changes, producing persistent trajectories.
Noise does not store memory of previous states.

```python
import random
import matplotlib.pyplot as plt

noise = []

for _ in range(100):
    noise.append(random.choice([-1,1]))

position = 0
walk = [position]

for _ in range(100):
    position += random.choice([-1,1])
    walk.append(position)

plt.figure(figsize=(8,4))

plt.plot(noise,label="Noise")
plt.plot(walk,label="Random Walk")

plt.legend()
plt.savefig("../images/Noise_and_RandomWalk.png", dpi=300)
plt.show()
```

# Oscillatory Systems and Mean Reversion

Feedback creates oscillatory behavior and repeated cycles using Mean Reversion which pulls the system back toward equilibrium.

```python
import random
import matplotlib.pyplot as plt

position = 0
velocity = 0

position_history = []

for _ in range(200):

    noise = random.choice([-1,1])

    force = -0.1*position - 0.05*velocity

    velocity += force + noise

    position += velocity

    position_history.append(position)

plt.figure(figsize=(10,5))

plt.plot(position_history)

plt.xlabel("Time")
plt.ylabel("Position")

plt.savefig("../images/Oscillator via MeanReversion.png", dpi=300)
plt.show()
```

```python
threshold = 1.0
settled_step = None

for i in range(len(position_history) - 1, -1, -1):
    if abs(position_history[i]) > threshold:
        settled_step = i + 1
        break

if settled_step is None:
    settled_step = 0

print(f"System settled within ±{threshold} of equilibrium by step {settled_step}")
```
